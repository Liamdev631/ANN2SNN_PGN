```python
import csv
import numpy as np

MODEL = "vgg16"
DATASET = "cifar100"

MODEL_CHECKPOINT = f"runs/{MODEL}-{DATASET}_log.txt"

epochs = []
train_loss = []
val_loss = []
val_acc = []
lr = []

with open(MODEL_CHECKPOINT) as f:
	reader = csv.reader(f, delimiter=',')
	next(reader, None) # skip header
	for row in reader:
		epochs.append(int(row[0].split(":")[1]))
		train_loss.append(float(row[1].split(":")[1]))
		val_loss.append(float(row[2].split(":")[1]))
		val_acc.append(float(row[3].split(":")[1]))
		lr.append(float(row[4].split(":")[1]))
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title("Log(Loss)")
plt.xlabel("Epoch")
plt.plot(epochs, np.log(train_loss), label='Train')
plt.plot(epochs, np.log(val_loss), label='Validation')
plt.legend()
plt.show()

plt.figure()
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.plot(epochs, val_acc)
plt.hlines(np.linspace(0, 1, 11)[1:-1], xmin=epochs[0], xmax=epochs[-1], linestyles="--", alpha=0.2)
plt.ylim(0, 1)
plt.show()

plt.figure()
plt.title("Learning Rate")
plt.xlabel("Epoch")
plt.plot(epochs, lr)
plt.show();
```

```python
print(f"Max Val Accuracy: {np.max(val_acc)}")
```
